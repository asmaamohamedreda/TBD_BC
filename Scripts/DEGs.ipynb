```r
metadata = read.csv("/content/metadata_GSE4290")
exp = read.csv("/content/exp_GSE4290", row.names = 1)
```

```r
exp_scaled_t = scale(t(exp))
exp_scaled = t(exp_scaled_t)
```

```r
design = model.matrix(~ 0 + Condition, data = metadata)
colnames(design) <- gsub("Condition", "", colnames(design))
design

contrast.matrix <- makeContrasts(
  Oligodendroglioma_vs_NonTumor = Oligodendroglioma - NonTumor,
  Astrocytoma_vs_NonTumor = Astrocytoma - NonTumor,
  Glioblastoma_vs_NonTumor = Glioblastoma - NonTumor,
  levels = design
)

fit <- lmFit(exp, design)
fit2 <- contrasts.fit(fit, contrast.matrix)
fit2 <- eBayes(fit2)
```

```r
results_Oligodendroglioma_vs_NonTumor <- topTable(fit2, coef = "Oligodendroglioma_vs_NonTumor", number = Inf,adjust = "BH")
results_Astrocytoma_vs_NonTumor <- topTable(fit2, coef = "Astrocytoma_vs_NonTumor", number = Inf,adjust = "BH")
results_Glioblastoma_vs_NonTumor <- topTable(fit2, coef = "Glioblastoma_vs_NonTumor", number = Inf,adjust = "BH")
```
